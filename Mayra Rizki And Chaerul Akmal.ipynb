###**Bagian 4.2 Jumlah Cluster (Elbow Method)**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(2, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    kmeans.fit(embeddings)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(2,11), wcss, marker='o')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

###**Bagian 4.3 Silhouette Score**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2,11):

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(embeddings)

    score = silhouette_score(
        embeddings,
        labels
    )

    print(f"k = {k}, Silhouette Score = {score:.4f}")
```

###**Bagian 5.3 Menampilkan Mention dan Hastag**

```python
import pandas as pd
import re

# Baca data
df = pd.read_csv("combined_tweets.csv")

# Pastikan kolom full_text tidak kosong
df = df.dropna(subset=["full_text"]).copy()

def format_tweet(row):
    text = str(row["full_text"]).lower()

    # Ambil username dari kolom username
    username = str(row["username"]).strip()

    # Jika username kosong / NaN
    if username == "nan" or username == "":
        username = "@user"
    elif not username.startswith("@"):
        username = "@" + username

    # Ambil semua hashtag dari tweet
    hashtags = re.findall(r"#\w+", text)

    # Hilangkan hashtag lama dari isi tweet
    text_clean = re.sub(r"#\w+", "", text)

    # Hilangkan mention lama dari isi tweet
    text_clean = re.sub(r"@\w+", "", text_clean)

    # Hilangkan URL
    text_clean = re.sub(r"http\S+|www\.\S+", "", text_clean)

    # Rapikan spasi
    text_clean = re.sub(r"\s+", " ", text_clean).strip()

    # Hilangkan hashtag duplikat
    hashtags = list(dict.fromkeys(hashtags))

    # Gabungkan format:
    # @username### #hashtag1 #hashtag2 isi tweet
    hasil = f"{username}###"

    if hashtags:
        hasil += " " + " ".join(hashtags)

    if text_clean:
        hasil += " " + text_clean

    return hasil

# Buat kolom baru
df["tweet_lda"] = df.apply(format_tweet, axis=1)

# Lihat hasil
print(df[["username", "full_text", "tweet_lda"]].head(10).to_string())

# Simpan hasil
df.to_csv(
    "combined_tweets_format_mention_hashtag.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nFile berhasil dibuat:")
print("combined_tweets_format_mention_hashtag.csv")
```

###**Bagian 5.4 Dictionary Corpus**

```python
import pandas as pd
from gensim import corpora
import ast # Import ast for literal_eval

# Reload df to ensure it contains the 'tokens' column from previous preprocessing steps
# This is crucial if a previous cell (like fjQpYvm4cbv1) overwrote 'df' with a different CSV.
df = pd.read_csv('hasil_preprocessing.csv')

# The 'tokens' column is often saved as a string representation of a list in CSV.
# Convert it back to a list of lists for gensim.corpora.Dictionary.
if not df.empty and 'tokens' in df.columns and isinstance(df['tokens'].iloc[0], str):
    df['tokens'] = df['tokens'].apply(ast.literal_eval)

dictionary = corpora.Dictionary(df["tokens"])

# Filter kata yang terlalu jarang atau terlalu sering muncul
dictionary.filter_extremes(
    no_below=2,
    no_above=0.80,
    keep_n=5000
)

corpus = [dictionary.doc2bow(tokens) for tokens in df["tokens"]]

print("\nJumlah kata unik setelah filter:", len(dictionary))
print("Jumlah dokumen untuk LDA:", len(corpus))
```

###**Bagian 5.5 Vektorisasi : Bag Of Words**

```python
print("\n" + "="*55)
print("BAGIAN 3: VEKTORISASI BAG OF WORDS")
print("="*55)
# PENTING: LDA menggunakan CountVectorizer (frekuensi kata mentah)
# BUKAN TF-IDF! TF-IDF akan mengacaukan asumsi probabilistik LDA
cv = CountVectorizer(
max_features = 300, # ambil 300 kata paling sering
min_df = 2, # kata harus muncul di min. 2 dokumen
max_df = 0.90, # abaikan kata yang ada di > 90% dokumen
ngram_range = (1, 1), # unigram saja untuk LDA
)
X = cv.fit_transform(df['clean_text'])
terms = cv.get_feature_names_out()

print(f"\n Vektorisasi selesai!")
print(f" Dimensi matriks BoW : {X.shape} (dokumen × kata)")
print(f" Jumlah fitur kata : {len(terms)}")
print(f" Top 30 kata paling sering:")
# Hitung frekuensi total setiap kata
word_freq = np.array(X.sum(axis=0)).flatten()
top30_idx = word_freq.argsort()[::-1][:30]
print(f" {', '.join([terms[i] for i in top30_idx])}")
```

###**Bagian 5.6 Mencari K Optimal - Elbow Perplexity**

```python
print("\n" + "="*55)
print("BAGIAN 4: MENCARI K OPTIMAL — ELBOW PERPLEXITY")
print("="*55)
K_range = range(2, 11)
perplexities = []
log_likelihoods = []
print("\n K | Perplexity | Log-Likelihood")
print(" " + "-"*38)
for k in K_range:
    lda_tmp = LatentDirichletAllocation(
        n_components = k,
        max_iter = 15,
        learning_method = 'online',
        random_state = 42
    )
    lda_tmp.fit(X)
    perp = lda_tmp.perplexity(X)
    logl = lda_tmp.score(X)
    perplexities.append(perp)
    log_likelihoods.append(logl)
    print(f" K={k:2d} | {perp:10.2f} | {logl:12.2f}")

best_k = list(K_range)[np.argmin(perplexities)]
print(f"\n K terbaik (perplexity terendah) = {best_k}")
# Plot Elbow
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
axes[0].plot(list(K_range), perplexities, 'o-', color='#7C3AED', lw=2.5, ms=9)
axes[0].axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'K\noptimal={best_k}')
axes[0].set_title('Perplexity vs K (↓ semakin baik)', fontsize=13,
fontweight='bold')
axes[0].set_xlabel('Jumlah Topik (K)')
axes[0].set_ylabel('Perplexity')
axes[0].legend(); axes[0].grid(alpha=0.3)
axes[1].plot(list(K_range), log_likelihoods, 's-', color='#0891B2', lw=2.5,
ms=9)
axes[1].axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'K\noptimal={best_k}')
axes[1].set_title('Log-Likelihood vs K (↑ semakin baik)', fontsize=13,
fontweight='bold')
axes[1].set_xlabel('Jumlah Topik (K)')
axes[1].set_ylabel('Log-Likelihood')
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.suptitle('Evaluasi LDA — Menentukan K Optimal | datacrawlmix.csv',
fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('01_lda_elbow.png', dpi=150, bbox_inches='tight')
plt.show()
print(" Grafik disimpan: 01_lda_elbow.png")
```

###**Bagian 5.7 Fit Model LDA Final**

```python
print("\n" + "="*55)
print("BAGIAN 5: FIT MODEL LDA FINAL")
print("="*55)
# bisa diganti n_components sesuai hasil evaluasi di atas
N_TOPICS = best_k # atau set manual, misal: N_TOPICS = 5

lda = LatentDirichletAllocation(
n_components = N_TOPICS,
max_iter = 30, # iterasi lebih banyak untuk konvergen
learning_method = 'online', # online variational Bayes (lebih cepat)
doc_topic_prior = None, # alpha = 1/K (default)
topic_word_prior= None, # beta = 1/K (default)
random_state = 42,
)
lda.fit(X)
print(f"\n LDA selesai dilatih!")
print(f" Jumlah topik : {lda.n_components}")
print(f" Log-likelihood : {lda.score(X):.2f}")
print(f" Perplexity : {lda.perplexity(X):.2f}")
```

**Hasil Topik LDA**

```python
print("\n" + "="*55)
print("BAGIAN 6: HASIL TOPIK LDA")
print("="*55)
N_TOP_WORDS = 15
print(f"\n{'='*60}")
print(f" TOPIK-TOPIK yang ditemukan LDA dari datacrawlmix.csv")
print(f"{'='*60}")
for topic_idx, topic_comp in enumerate(lda.components_):
    top_indices = topic_comp.argsort()[::-1][:N_TOP_WORDS]
    top_words = [terms[i] for i in top_indices]
    top_weights = [topic_comp[i] for i in top_indices]
    print(f"\n TOPIK {topic_idx}:")
    print(f" Top kata : {', '.join(top_words[:10])}")
    print(f" Bobot : {', '.join([f'{w:.1f}' for w in top_weights[:10]])}")
```

**Assign Topik Setiap Tweet**

```python
print("\n" + "="*55)
print("BAGIAN 7: ASSIGN TOPIK KE SETIAP TWEET")
print("="*55)
doc_topic_matrix = lda.transform(X) # shape: (n_docs, n_topics)
df['dominant_topic'] = doc_topic_matrix.argmax(axis=1)
df['topic_prob'] = doc_topic_matrix.max(axis=1)
print(f"\n Distribusi topik dominan:")
print(df['dominant_topic'].value_counts().sort_index().to_string())
print(f"\n Contoh tweet per topik:")
for k in range(N_TOPICS):
    subset = df[df['dominant_topic'] == k].sort_values('topic_prob',
        ascending=False)
    print(f"\n ── Topik {k} (n={len(subset)}, avg_prob={subset['topic_prob'].mean():.3f}) ──")
    for _, row in subset.head(2).iterrows():
        print(f" → [{row['topic_prob']:.2f}] {str(row['full_text'])[:85]}...")
```

```python
# Make a copy to avoid modifying the original DataFrame if it's used elsewhere
df_cleaned = df.copy()

# Fill NaN values in 'username' with a placeholder
# This is crucial for NetworkX to treat these as valid, distinct nodes if they represent sources.
# For mentions, we can filter them out later if they are the target.
df_cleaned['username'] = df_cleaned['username'].fillna('anonymous_user')

# Filter out rows where 'full_text' might be NaN, which could lead to issues
df_cleaned = df_cleaned.dropna(subset=['full_text'])

edges_cleaned = []
for i, row in df_cleaned.iterrows():
    source_user = str(row["username"])

    # FIX: Corrected regex to extract mentions (removed extra backslash)
    mentions = [m for m in re.findall(r"@(\w+)", row["full_text"]) if m]

    for mention in mentions:
        # FIX: Removed condition 'source_user != 'anonymous_user'' to allow anonymous users to be sources
        # Only add an edge if the target mention is valid (not empty)
        if mention != '':
            edges_cleaned.append((source_user, mention))

print(f"Jumlah relasi mention setelah pembersihan: {len(edges_cleaned)}")
print("Contoh relasi mention setelah pembersihan:")
print(edges_cleaned[:5])

# Rebuild the graph with cleaned edges
G_cleaned = nx.DiGraph()
G_cleaned.add_edges_from(edges_cleaned)

# Calculate centrality metrics for the cleaned graph
degree_centrality_cleaned = nx.degree_centrality(G_cleaned)
betweenness_centrality_cleaned = nx.betweenness_centrality(G_cleaned)

try:
    eigenvector_centrality_cleaned = nx.eigenvector_centrality(G_cleaned)
except nx.PowerIterationFailedConvergence:
    eigenvector_centrality_cleaned = {}

print("\n--- Centrality Metrics (After Cleaning) ---")
print("Top pengguna (Degree Centrality):")
# Convert node names to string for consistent display and sort
print(sorted([(str(k), v) for k, v in degree_centrality_cleaned.items()], key=lambda x: x[1], reverse=True)[:5])

print("\nTop pengguna (Betweenness Centrality):")
print(sorted([(str(k), v) for k, v in betweenness_centrality_cleaned.items()], key=lambda x: x[1], reverse=True)[:5])

print("\nTop pengguna (Eigenvector Centrality):")
print(sorted([(str(k), v) for k, v in eigenvector_centrality_cleaned.items()], key=lambda x: x[1], reverse=True)[:5])
```

**Edge List**

```python
edges_rows = []

# Ensure G is defined and not empty before iterating over its edges
# If G is empty (e.g., due to previous errors), this loop won't run, and edges_rows will remain empty.
# Check if G is in locals() and if it has any edges.
if 'G' in locals() and not G.edges():
    print("Graph 'G' is empty. No edges to process for Edges_Demo.")

# Populate edges_rows if G has edges
if 'G' in locals() and G.edges():
    for u, v, d in G.edges(data=True):
        # Ensure node2comm is defined, if not, assume default -1 or similar
        community_source = node2comm.get(u, -1) if 'node2comm' in locals() else -1
        community_target = node2comm.get(v, -1) if 'node2comm' in locals() else -1

        edges_rows.append({
            "Source": u,
            "Target": v,
            "Type": "Directed",
            "Weight": d.get("weight", 1),
            "Community_Source": community_source,
            "Community_Target": community_target,
            "Intra_Community": community_source == node2comm.get(v, -2) if 'node2comm' in locals() else False,
        })

# Initialize Edges_Demo with predefined columns even if edges_rows is empty
# This prevents KeyError if there are no edges and sort_values is called on an empty DataFrame.
if not edges_rows:
    Edges_Demo = pd.DataFrame(columns=["Source", "Target", "Type", "Weight", "Community_Source", "Community_Target", "Intra_Community"])
    print("Edges_Demo DataFrame initialized as empty with defined columns.")
else:
    Edges_Demo = (pd.DataFrame(edges_rows)
                  .sort_values("Weight", ascending=False)
                  .reset_index(drop=True))

# Display the head of Edges_Demo to confirm creation
print("Edges_Demo DataFrame created successfully:")
print(Edges_Demo.head())
```

#### **Bagian 7.1 Import, MinSup, FTC**

```python
from collections import defaultdict
import itertools
import pandas as pd # Import pandas if not already imported globally

def FTC(D, minsup):
    """
    Translates the pseudo-code for Frequent Term Set Clustering (FTC) into Python.
    This implementation now includes concrete (though simplified) implementations
    for placeholder functions.

    Args:
        D: The database (expected to be a list of lists of tokens, e.g., df['tokens']).
        minsup: Minimum support threshold (float, e.g., 0.05 for 5% support).

    Returns:
        A tuple containing:
        - SelectedTermSets: A set of term sets selected to cover the database.
        - total_coverage: A set of document IDs covered by SelectedTermSets.
    """

    SelectedTermSets = set()
    n = len(D)  # Total number of documents in D (translated from: n:= |D|;)
    min_count = int(minsup * n) # Convert minsup from fraction to minimum count

    # --- Helper Functions Implementation ---
    def DetermineFrequentTermsets(database_tokens, min_support_count):
        # This function identifies frequent single terms and pairs of terms
        # based on the min_support_count.
        print("  [DEBUG] Menentukan Frequent Termsets...")
        term_counts = defaultdict(int)
        frequent_term_sets = set()

        # Count individual terms and pairs of terms in each document
        for doc_id, doc_tokens in enumerate(database_tokens):
            unique_doc_tokens = list(set(doc_tokens)) # Count only once per document

            # Count single terms
            for term in unique_doc_tokens:
                term_counts[frozenset({term})] += 1

            # Count pairs of terms
            for pair in itertools.combinations(unique_doc_tokens, 2):
                term_counts[frozenset(pair)] += 1

        # Filter based on min_support_count
        for term_set, count in term_counts.items():
            if count >= min_support_count:
                frequent_term_sets.add(term_set)

        print(f"  [DEBUG] Ditemukan {len(frequent_term_sets)} frequent term sets.")
        return frequent_term_sets

    def get_documents_covered_by_term_set(term_set_to_check, D_original_tokens):
        # This function returns a set of document IDs from D_original_tokens
        # that are covered by 'term_set_to_check'.
        covered_docs = set()
        for doc_id, doc_tokens in enumerate(D_original_tokens):
            # A document is covered if it contains ALL terms in the term_set_to_check
            if term_set_to_check.issubset(set(doc_tokens)):
                covered_docs.add(doc_id)
        return covered_docs

    def calculate_total_coverage(D_original_tokens, current_selected_term_sets):
        # This function calculates the union of all documents covered by
        # the term sets in 'current_selected_term_sets'.
        total_covered = set()
        for ts in current_selected_term_sets:
            total_covered.update(get_documents_covered_by_term_set(ts, D_original_tokens))
        return total_covered
    # ---------------------------------------------------------------

    RemainingTermSets = DetermineFrequentTermsets(D, min_count)

    # Track documents that are not yet covered
    uncovered_documents = set(range(n))  # Assuming D has n documents indexed 0 to n-1

    while len(uncovered_documents) != 0:
        if not RemainingTermSets:
            print("  [WARNING] Tidak ada lagi kandidat term set yang bisa dipertimbangkan, tetapi belum semua dokumen tercakup.")
            break

        best_candidate = None
        max_new_coverage = -1

        for candidate_set in RemainingTermSets:
            candidate_coverage = get_documents_covered_by_term_set(candidate_set, D)
            new_coverage_for_candidate = len(candidate_coverage.intersection(uncovered_documents))

            if new_coverage_for_candidate > max_new_coverage:
                max_new_coverage = new_coverage_for_candidate
                best_candidate = candidate_set

        if best_candidate is None or max_new_coverage == 0:
            print("  [WARNING] Tidak dapat menemukan kandidat yang mencakup dokumen baru. Algoritma berhenti.")
            break

        SelectedTermSets.add(best_candidate)
        RemainingTermSets.remove(best_candidate)

        covered_by_best_candidate = get_documents_covered_by_term_set(best_candidate, D)
        uncovered_documents.difference_update(covered_by_best_candidate)

    total_coverage = calculate_total_coverage(D, SelectedTermSets)
    return SelectedTermSets, total_coverage

# --- Contoh Penggunaan --- (Ini dapat dihapus setelah pengujian)
# Gunakan df['tokens'] dari notebook Anda sebagai input D
# Pastikan df sudah dimuat dan kolom 'tokens' sudah ada.

# Memuat df jika belum ada (misal, setelah runtime reset)
try:
    # Assuming df is already loaded in the notebook from previous cells
    if 'df' not in locals() or 'tokens' not in df.columns:
        print("[INFO] Memuat kembali df dan tokens untuk contoh penggunaan.")
        df = pd.read_csv('hasil_preprocessing.csv')
        if 'tokens' not in df.columns:
            # If 'tokens' column is missing, try to re-tokenize if 'clean_text' is available
            from nltk.tokenize import word_tokenize
            df['tokens'] = df['clean_text'].apply(lambda x: word_tokenize(x))
        print("[INFO] df dan tokens berhasil dimuat/dibuat kembali.")

    if not df.empty and 'tokens' in df.columns:
        actual_D = df['tokens'].tolist()
        actual_minsup = 0.05 # Contoh: 5% minimum support

        print("\n--- Menjalankan Algoritma FTC --- (Dengan data dari notebook Anda)")
        selected_sets, covered_docs = FTC(actual_D, actual_minsup)

        print("\n--- Hasil Algoritma FTC ---")
        print("Selected Term Sets:")
        for ts in selected_sets:
            print(f"  - {set(ts)}") # Convert frozenset back to set for display
        print(f"Total Documents Covered (IDs): {len(covered_docs)} dari {len(actual_D)} dokumen")
        # print("List ID Dokumen yang Tercakup:", covered_docs) # uncomment to see document IDs
    else:
        print("[WARNING] df atau kolom 'tokens' tidak tersedia. Contoh penggunaan FTC dilewati.")
except Exception as e:
    print(f"[ERROR] Terjadi kesalahan saat menjalankan contoh FTC: {e}")
    print("Pastikan 'hasil_preprocessing.csv' tersedia dan `df['tokens']` telah dibuat.")
```

####**Bagian 7.2 Import Counter**

```python
from collections import Counter
import ast # Untuk mengonversi string list menjadi list

all_terms = []

# Periksa tipe data di df['tokens']
# Jika berupa string, ubah menjadi list menggunakan ast.literal_eval
# Jika sudah list, langsung gunakan
if isinstance(df['tokens'].iloc[0], str):
    # Mengonversi string representasi list menjadi list sebenarnya
    df['tokens'] = df['tokens'].apply(ast.literal_eval)

for tokens in df["tokens"]:
    all_terms.extend(tokens)

term_frequency = Counter(all_terms)

df_frequency = pd.DataFrame(
    term_frequency.items(),
    columns=["term", "frequency"]
)

df_frequency = df_frequency.sort_values(
    by="frequency",
    ascending=False
).reset_index(drop=True)

print("\nTop 20 term paling sering muncul:")
print(df_frequency.head(20))
```

####**Bagian 7.3 Menghitung Entropy**

```python
import math # Import the math module
total_documents = len(df)

term_entropy = {}

for term in term_frequency.keys():

    # Tweet yang memuat term tersebut
    doc_indices = []

    for index, tokens in enumerate(df["tokens"]):
        if term in tokens:
            doc_indices.append(index)

    jumlah_muncul = len(doc_indices)

    # Karena setiap tweet dianggap satu kemunculan term,
    # probabilitasnya seragam pada tweet yang memuat term.
    if jumlah_muncul > 0:
        probability = 1 / jumlah_muncul
        entropy = -jumlah_muncul * probability * math.log2(probability)
    else:
        entropy = 0

    term_entropy[term] = entropy

df_entropy = pd.DataFrame(
    term_entropy.items(),
    columns=["term", "entropy"]
)

# Gabungkan frekuensi + entropy
df_term = pd.merge(
    df_frequency,
    df_entropy,
    on="term",
    how="left"
)

df_term = df_term.sort_values(
    by=["frequency", "entropy"],
    ascending=[False, True]
).reset_index(drop=True)

print("\nTop term dengan frequency dan entropy:")
print(df_term.head(20))
```

```python
minimum_frequency = 3

frequent_terms = df_term[
    df_term["frequency"] >= minimum_frequency
].copy()

print("\nJumlah frequent terms:", len(frequent_terms))
```

```python
from itertools import combinations # Import combinations for pairwise term processing

term_documents = {}

for term in frequent_terms["term"]:
    term_documents[term] = set(
        df.index[
            df["tokens"].apply(lambda tokens: term in tokens)
        ]
    )

overlap_results = []

frequent_term_list = frequent_terms["term"].tolist()

for term_a, term_b in combinations(frequent_term_list, 2):

    docs_a = term_documents[term_a]
    docs_b = term_documents[term_b]
    intersection = len(docs_a.intersection(docs_b)) # Corrected indentation
    min_docs = min(len(docs_a), len(docs_b))

    if min_docs > 0:
        overlap_score = intersection / min_docs
    else:
        overlap_score = 0

    overlap_results.append({
        "term_1": term_a,
        "term_2": term_b,
        "frequency_term_1": len(docs_a),
        "frequency_term_2": len(docs_b),
        "co_occurrence": intersection,
        "overlap": round(overlap_score, 4)
    })

df_overlap = pd.DataFrame(overlap_results)

df_overlap = df_overlap.sort_values(
    by="overlap",
    ascending=False
).reset_index(drop=True)

print("\nTop 20 pasangan term dengan overlap tertinggi:")
print(df_overlap.head(20))
```

####**Bagian 7.4 Overlap Entropy**

```python
overlap_threshold = 0.50

clusters = []
visited_terms = set()

for term in frequent_term_list:

    if term in visited_terms:
        continue

    cluster = {term}
    terms_to_check = [term]

    while terms_to_check:
        current_term = terms_to_check.pop()
        related_terms = df_overlap[
            (
                (df_overlap["term_1"] == current_term) |
                (df_overlap["term_2"] == current_term)
            )
            &
            (df_overlap["overlap"] >= overlap_threshold)
        ]

        for _, row in related_terms.iterrows():

            if row["term_1"] == current_term:
                related_term = row["term_2"]
            else:
                related_term = row["term_1"]

            if related_term not in cluster:
                cluster.add(related_term)
                terms_to_check.append(related_term)

    visited_terms.update(cluster)
    clusters.append(sorted(cluster))
```

```python
from collections import Counter
import pandas as pd
import ast # To convert string list to actual list

# Load df to ensure it's defined and contains the 'tokens' column
df = pd.read_csv('/content/hasil_preprocessing.csv')

# Convert 'tokens' column from string representation to actual list objects if it's not already
if not df.empty and 'tokens' in df.columns and isinstance(df['tokens'].iloc[0], str):
    df['tokens'] = df['tokens'].apply(ast.literal_eval)

min_support = 2

counter = Counter()

# Use 'tokens' column instead of 'Words'
for doc in df["tokens"]:
    counter.update(set(doc))

frequent_terms = {
    term
    for term, freq in counter.items()
    if freq >= min_support
}

print("Jumlah Frequent Term :", len(frequent_terms))

df["Frequent_Term_Set"] = df["tokens"].apply(
    lambda words: sorted(list(set(words) & frequent_terms))
)
```

####**Bagian 7.6 Entropy Overlap**

```python
from collections import Counter
import numpy as np # Import numpy for log2

# Create a mapping from original conversation_id_str to D1, D2, ...
unique_conversation_ids = df['conversation_id_str'].unique()
doc_id_map = {cid: f"D{i+1}" for i, cid in enumerate(unique_conversation_ids)}

rows = []

for _, row in df.iterrows():

    # Use the mapped document ID
    doc_id_original = row["conversation_id_str"]
    doc_id_formatted = doc_id_map[doc_id_original]

    words = row["tokens"] # Corrected column name

    counts = Counter(words)
    total = len(words)

    for word in words:

        # Frequent Term Set
        if word in frequent_terms:
            ft = "Ya"
        else:
            ft = "Tidak"

        # Entropy Overlap
        # Handle case where p might be 0 to avoid log2(0) error
        if total > 0 and counts[word] > 0:
            p = counts[word] / total
            entropy = -(p * np.log2(p))
        else:
            entropy = 0.0 # If word count is 0 or total is 0, entropy is 0

        rows.append({
            "Words": word,
            "Document_ID": doc_id_formatted,
            "Frequent_Term_Set": ft, # Re-added this column
            "Entropy_Overlap": round(entropy, 4)
        })

hasil = pd.DataFrame(rows)
```
